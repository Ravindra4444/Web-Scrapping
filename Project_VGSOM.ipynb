```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
page = requests.get("https://www.naukri.com/jobs-in-kolkata", headers=headers)
response = BeautifulSoup(page.content, 'html.parser')
print(response.prettify())
```

```python
job_containers = response.find_all('span', class_="content")
print(type(job_containers))
print(len(job_containers))
```

```python
salary_containers = response.find_all('div', class_="other_details")
print(type(salary_containers))
print(len(salary_containers))
```

```python
first_salary = salary_containers[0]
first_salary.find('span', class_= 'salary').text
```

```python
profile = []
name = []
exp = []
location = []
skills = []
salary = []
for containers in job_containers:
        profiles = containers.ul.li.a.text
        profile.append(profiles)
        names = containers.find('span', class_= 'org').text
        name.append(names)
        experience = containers.find('span', class_= 'exp').text
        exp.append(experience)
        loc = containers.find('span', class_= 'loc').text
        location.append(loc)
        skill = containers.find('span', class_= 'skill').text
        skills.append(skill)
len(skills)
```

```python
for sal in salary_containers:
    salar = sal.find('span', class_= 'salary').text
    salary.append(salar)
len(salary)
```

```python
import pandas as pd

test_df = pd.DataFrame({'Job Profile': profile,
                       'Company Name': name,
                       'Experience': exp,
                       'Location': location,
                       'Skills Required': skills,
                       'Salary': salary})
print(test_df.info())
test_df
```
